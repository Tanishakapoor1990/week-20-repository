```python
#Importing libraries 
from nltk.corpus import gutenberg

#Checking fileids in corpus gutenberg
print(gutenberg.fileids())
```

```python
# Getting words from corpus gutenberg
paradise_lost_words = gutenberg.words('milton-paradise.txt')
print(paradise_lost_words)

#Getting sentences from corpus gutenberg
paradise_lost_sents = gutenberg.sents('milton-paradise.txt')[:10]
#print(len(paradise_lost_sents))
```

```python
#importing libraries 
from string import punctuation 
print(punctuation)

without_punct = [word for word in paradise_lost_words if word not in punctuation]
print(without_punct)
```

```python
from nltk.corpus import stopwords

sw = stopwords.words('english')
print(sw)
```

```python
without_sw = [word for word in without_punct if word.lower() not in sw]
without_sw
```

```python
import nltk
freqdist= nltk.FreqDist(without_sw)
freqdist
```

```python
print(len(freqdist))
```

```python
print(freqdist.most_common(20))
```

```python
# Frequency Distribution Plot
import matplotlib.pyplot as plt
freqdist.plot(30,cumulative=False)
plt.show()
```

```python
# top most words by frequency:
top_20=freqdist.most_common(20)

listnew = [key for key, val in top_20 for _ in range(val+1)]

plt.hist(listnew, bins=20)
plt.xticks(rotation=90)
plt.show()
```

```python
#Importing porterstemmer library
from nltk.stem import PorterStemmer

#Creating porterstemmer object
ps = PorterStemmer()

#Looping through the loop 
for i in without_sw :
    print("Stem for {} is {}".format(i, ps.stem(i))) 
```

```python
#importing libraries
import nltk
from nltk.stem import WordNetLemmatizer

#Creating wornetlemmatizer object
wordnet_lemmatizer = WordNetLemmatizer()

for w in without_sw:
    print("Lemma for {} is {}".format(w, wordnet_lemmatizer.lemmatize(w))) 
```

In my opinion lemmatization is better as it  usually refers to doing things properly with the use of a vocabulary 
and morphological analysis of words, normally aiming to remove inflectional endings only and to return
the base or dictionary form of a word, which is known as the lemma .

Stemming usually refers to a crude heuristic process that chops off the ends 
of words in the hope of achieving this goal correctly most of the time, and often includes the removal 
of derivational affixes.

Are the sentences and their sentiment analysis scores correct? Explain why or why not. 

```python
#Checking sentences from corpus gutenberg
from nltk.corpus import gutenberg
paradise_lost_sents = gutenberg.sents('milton-paradise.txt')

print(len(paradise_lost_sents))
```

```python
#Writing a function to concatenate the data back into sentences with one readable sentence in each list.
def to_list_sentences(paradise_lost_sents):
    #takes in a list of lists
    sentence_list = []
    #for each list 
    for sentence in paradise_lost_sents:
        sentence_string =''
        # for each word in the list
        for word in sentence:
            sentence_string += ' ' + word
        temp = []
        temp.append(sentence_string)
        sentence_list.append(temp)
    return sentence_list
    
sents_analysed=to_list_sentences(paradise_lost_sents)

```

```python
#imprting library 
from nltk.sentiment.vader import SentimentIntensityAnalyzer
#nltk.download('vader_lexicon')

#creating object function
analyzer = SentimentIntensityAnalyzer()
```

```python
# Generating sentiment for all the sentence present in the dataset

import pandas as pd
sentiments=[]



for sentence in sents_analysed:
    #print(sentence[0])
    sentiment = analyzer.polarity_scores(sentence[0])
    #print(sentiment)
    compound = sentiment["compound"]
    pos = sentiment["pos"]
    neu = sentiment["neu"]
    neg = sentiment["neg"]
    sentiments.append({
                    "text":sentence,
                    "compound":compound,
                    "positive":pos,
                    "negative":neg,
                    "neutral":neu
})
#sentiments
df = pd.DataFrame(sentiments)
df
```

```python
# Top 5 neutral sentences:
df.sort_values(by= ['neutral'],ascending=False)[:5]
```

```python
# Top 5 positive lines:
df.sort_values(by= ['positive'],ascending=False)[:5]
```

```python
# Top 5 negative lines:
df.sort_values(by= ['negative'],ascending=False)[:5]
```
